```python
import tensorflow as tf
tf.keras.backend.clear_session()
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(filters = 64, kernel_size = (1, 5), strides = (1, 1), 
                           activation=None, use_bias = False, kernel_initializer='he_uniform'), 
    tf.keras.layers.LeakyReLU(alpha=0.2),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.SpatialDropout2D(rate=0.2),
    tf.keras.layers.Conv2D(filters = 64, kernel_size = (1, 5), strides = (1, 1), 
                           activation=None, use_bias = False, kernel_initializer='he_uniform'),
    tf.keras.layers.LeakyReLU(alpha=0.2),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.SpatialDropout2D(rate=0.2),
    tf.keras.layers.Conv2D(filters = 64, kernel_size = (1, 5), strides = (1, 2), 
                           activation=None, use_bias = False, kernel_initializer='he_uniform'),
    tf.keras.layers.LeakyReLU(alpha=0.2),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.SpatialDropout2D(rate=0.2),
    tf.keras.layers.Conv2D(filters = 128, kernel_size = (1, 5), strides = (1, 1), 
                           activation=None, use_bias = False, kernel_initializer='he_uniform'),
    tf.keras.layers.LeakyReLU(alpha=0.2),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.SpatialDropout2D(rate=0.2),
    tf.keras.layers.Conv2D(filters = 128, kernel_size = (1, 5), strides = (1, 1), 
                           activation=None, use_bias = False, kernel_initializer='he_uniform'),
    tf.keras.layers.LeakyReLU(alpha=0.2),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.SpatialDropout2D(rate=0.2),
    tf.keras.layers.Conv2D(filters = 128, kernel_size = (1, 5), strides = (1, 2), 
                           activation=None, use_bias = False, kernel_initializer='he_uniform'),
    tf.keras.layers.LeakyReLU(alpha=0.2),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.SpatialDropout2D(rate=0.2),

    tf.keras.layers.Conv2D(filters = 256, kernel_size = (1, 5), strides = (1, 1), 
                           activation=None, use_bias = False, kernel_initializer='he_uniform'),
    tf.keras.layers.LeakyReLU(alpha=0.2),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.SpatialDropout2D(rate=0.2),
    tf.keras.layers.Conv2D(filters = 256, kernel_size = (1, 5), strides = (1, 1), 
                        activation=None, use_bias = False, kernel_initializer='he_uniform'),
    tf.keras.layers.LeakyReLU(alpha=0.2),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.SpatialDropout2D(rate=0.2),
    tf.keras.layers.Conv2D(filters = 256, kernel_size = (1, 5), strides = (1, 2), 
                           activation=None, use_bias = False, kernel_initializer='he_uniform'),
    tf.keras.layers.LeakyReLU(alpha=0.2),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.SpatialDropout2D(rate=0.2),

    tf.keras.layers.Conv2D(filters = 512, kernel_size = (1, 5), strides = (1, 1), 
                           activation=None, use_bias = False, kernel_initializer='he_uniform'),
    tf.keras.layers.LeakyReLU(alpha=0.2),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.SpatialDropout2D(rate=0.2),
    tf.keras.layers.Conv2D(filters = 512, kernel_size = (1, 5), strides = (1, 1), 
                        activation=None, use_bias = False, kernel_initializer='he_uniform'),
    tf.keras.layers.LeakyReLU(alpha=0.2),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.SpatialDropout2D(rate=0.2),
    tf.keras.layers.Conv2D(filters = 512, kernel_size = (1, 5), strides = (1, 2), 
                           activation=None, use_bias = False, kernel_initializer='he_uniform'),
    tf.keras.layers.LeakyReLU(alpha=0.2),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.SpatialDropout2D(rate=0.2),

    tf.keras.layers.Conv2D(filters = 1024, kernel_size = (1, 5), strides = (1, 1), padding='same',
                         activation=None, use_bias = False, kernel_initializer='he_uniform'),
    tf.keras.layers.LeakyReLU(alpha=0.2),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.SpatialDropout2D(rate=0.2),
    tf.keras.layers.Conv2D(filters = 1024, kernel_size = (1, 5), strides = (1, 1), padding='same', 
                        activation=None, use_bias = False, kernel_initializer='he_uniform'),
    tf.keras.layers.LeakyReLU(alpha=0.2),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.SpatialDropout2D(rate=0.2),
    tf.keras.layers.Conv2D(filters = 1024, kernel_size = (1, 5), strides = (1, 2), padding='same',
                           activation=None, use_bias = False, kernel_initializer='he_uniform'),
    tf.keras.layers.LeakyReLU(alpha=0.2),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.SpatialDropout2D(rate=0.2),


    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(units=20, activation=None, use_bias=False, name = 'compress'),
    tf.keras.layers.LeakyReLU(alpha=0.2),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(units=2, activation = None, use_bias=False) 
])
```
